```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

#**NumPy**  

## Answer - 01

```python
# Q1. Create NumPy arrays from the ages and salaries data generated above. Print both arrays and display their dtype, ndim, shape, and size.

ages_arr = np.array(ages)
salaries_arr = np.array(salaries)

print("Age Information of Employees:")
print("Age Array:",ages_arr)
print("Type:", ages_arr.dtype)
print("Dimention:", ages_arr.ndim)
print("Shape:", ages_arr.shape)
print("Size:", ages_arr.size)
print("\n")
print("Salary Information of Employees:")
print("Salary Array:",salaries_arr)
print("Type:", salaries_arr.dtype)
print("Dimention:", salaries_arr.ndim)
print("Shape:", salaries_arr.shape)
print("Size:", salaries_arr.size)
```

## Answer - 02

```python
# Q2. Using the salaries array, find and print the highest and lowest salary values. Also, calculate and print the average salary and age using NumPy functions.

print("Highest Salary:",max(salaries_arr))
print("Lowest Salary:",min(salaries_arr),"\n")

print("Average Salary:",np.average(salaries_arr))
print("Age Average:",np.average(ages_arr))
```

## Answer - 03

```python
# Q3. From the ages array, filter and print all ages greater than 30 using a boolean condition. Then count and print how many employees are older than 30.

bool_con_arr = []

for age in ages:
  if age > 30:
    bool_con_arr.append(True)
  else:
    bool_con_arr.append(False)

print(f"Boolean Mask: {bool_con_arr}")
print(f"Age above 30: {ages[bool_con_arr]}")
print(f"Count Above 30: {(ages_arr[bool_con_arr].size)}")
```

## Answer - 04

```python
# Q4. Create a new NumPy array that increases every employeeâ€™s age by 5 years (without modifying the original array). Print the new updated ages array.

broadcasted_arr = ages + 5
print(f"After increasing age by 5 years: {broadcasted_arr}")
print(f"Original Array: {ages_arr}")
```

## Answer - 05

```python
# Q5. Using NumPy, calculate the total salary expense (sum of all salaries) and the difference between the maximum and minimum salary. Print both results.

total = np.sum(salaries_arr)
difference = np.max(salaries_arr) - np.min(salaries_arr)

print(f"Total Salary Expense: {total}")
print(f"Salary Difference: {difference}")
```

```python
# Q6. Load the employees.csv file into a Pandas DataFrame. Display the first 5 rows, and check basic info using info() and summary statistics using describe().

from IPython.display import display
DF = pd.read_csv("employees.csv")

print("The DataFrame:")
display(DF)
print("\n\nThe first 5 rows:")
display(DF.head())
print("\n\nBasic information of DataFrame:")
DF.info()
print("\n\nSummary Statistics of DataFrame:")
display(DF.describe())
```

## **Answer - 07**

```python
# Q7. Display only the id, age, and salary columns. Then show the last 3 rows using tail().

col = ["id","age","salary"]
print("DataFrame with 3 columns:\n")
display(DF[col])

print("\n\nThe last 3 rows:\n")
display(DF[col].tail(3))
```

## **Answer - 08**

```python
# Q8. Filter the DataFrame to show only employees who work in the IT department. Print the result and show the total number of IT employees.

it_employees = DF[DF['dept'] == 'IT']
print(f"DataFrame with IT employees:\n")
display(it_employees)
print("\n")
print(f"\nTotal number of IT employees: {len(it_employees)}")
```

## **Answer - 09**

```python
# Q9. Sort the DataFrame by salary in descending order and display the top 3 highest-paid employees along with their department and age.

sorted_df = DF.sort_values(by='salary', ascending=False)
print("DataFrame sorted by salary in descending order:\n")
display(sorted_df)

print("\n\nTop 3 highest-paid employees (department and age):\n")
display(sorted_df.head(3)[['salary', 'dept', 'age']])
```

```python
# Q10. Replace all salary values greater than 80000 with 80000 using loc. Then calculate and print the new average salary of all employees after replacement.

# Replace all salaries greater than 80000 with 80000
DF.loc[DF['salary'] > 80000, 'salary'] = 80000

def highlight_80000(s):
    return ['background-color: gray' if v == 80000 else '' for v in s]

# Apply styling to highlight salary == 80000
styler = DF.style.apply(highlight_80000, subset=['salary'])
display(styler)

new_avg_salary = DF['salary'].mean()
print(f"\nNew average salary of all employees: {new_avg_salary}")
```
